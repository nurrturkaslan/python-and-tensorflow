```python
y = data["Type"].values
X = data.drop("Type", axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# EarlyStopping ve Dropout u overfitting problemini çözmek için kullanacağız
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy",optimizer = "adam")
```

```python
model.fit(X_train, y_train, epochs =700 , validation_data = (X_test, y_test), verbose = 1)
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot();
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy",optimizer = "adam")
```

```python
earlyStopping = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 25)
```

```python
model.fit(X_train, y_train, epochs = 700, validation_data= (X_test, y_test), verbose = 1, callbacks = [earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot();
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy",optimizer = "adam")
```

```python
model.fit(X_train, y_train, epochs = 700, validation_data= (X_test, y_test), verbose = 1, callbacks = [earlyStopping])
```

```python
kayipDf =pd.DataFrame(model.history.history)
kayipDf.plot()
```

```python
tahminlerimiz = model.predict_classes(X_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, tahminlerimiz))
```

```python
print(confusion_matrix(y_test, tahminlerimiz))
```
