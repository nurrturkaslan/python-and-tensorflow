```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#yazdığınız grafiklerin kolayca gösterilmesine olanak tanır
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,90,85,80,70,75,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "r")
plt.xlabel("yaş")
plt.ylabel("kilo")
plt.title("Kilo'nun Yaş'a göre değişimi")
plt.show()
```

```python
numpyDizi1 = np.linspace(0,10,20)
numpyDizi1
```

```python
numpyDizi2 = numpyDizi1 * 2
numpyDizi2
```

```python
plt.plot(numpyDizi1, numpyDizi2, "g--")
```

```python
plt.plot(numpyDizi1, numpyDizi2, "m*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizi1, numpyDizi2, "r*-")
plt.subplot(1,2,2)# 1sıra 2 kolan 2.grafik manasına geliyor
plt.plot(numpyDizi2, numpyDizi1, "m--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.8, 0.8, 0.3, 0.3])
figureAxes.plot(numpyDizi1, numpyDizi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik başlığı")
plt.show()
```

```python
#İçe İçe geçmiş garafikler
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.6,0.3,0.3])

eksen1.plot(numpyDizi1, numpyDizi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizi2, numpyDizi1, "b")
eksen2.set_xlabel("X Ekseni küçük ")
eksen2.set_ylabel("Y Ekseni küçük")
eksen2.set_title("Küçük Grafik Başlığı")
plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows = 1, ncols = 2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizi1, numpyDizi2, "g")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()
```

```python
yeniFigur = plt.figure(figsize = (4,4), dpi =100)

yenieksen = yeniFigur.add_axes([0.1, 0.1, 0.9, 0.9])

yenieksen.plot(numpyDizi1, numpyDizi2, label ="numpy dizi ** 2")
yenieksen.plot(numpyDizi1, numpyDizi1 ** 3, label ="numpy dizi ** 3")
yenieksen.legend(loc=6)
```

```python
yeniFigur.savefig("benimFigur.png", dpi = 200)
```
